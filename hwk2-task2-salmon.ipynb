```python
# FINAL CODE
import pandas as pd

'''
Read in AK Fish & Game 1878 - 1997 commercial salmon catch data.
Drop Notes and Comments column
'''
catch_data = (pd.read_csv('data/salmon_data.csv')
                .drop(columns = 'notesRegCode')
)

# Change the `Catch` column to be type `int64`
catch_data.at[400, 'Catch'] = 1
catch_data['Catch'] = catch_data['Catch'].astype('int64')

# Plot average salmon catches by management region
(catch_data.groupby('Region')['Catch']
            .mean()
            .sort_values(ascending = False)
            .plot(kind = 'bar',
                  title = "Alaska average commerical salmon catches by management region from 1878 to 1997",
                  ylabel = "Commercial catches of salmon (in thousands of fish)")
)            
```
